## Designing the architecture of the Cycle GANs

__Starting with Generator__

![Alt text](../images/generator_architecture.png?raw=true "Generator Architecture")

```python
#designing the residual block

class ResidualBlock(nn.Module):

    def __init__(self, in_channels):
        super(ResidualBlock, self).__init__()

        self.conv_block = nn.Sequential(
            nn.Conv2d(in_channels, in_channels, kernel_size=3, padding=1, bias=False),
            nn.InstanceNorm2d(in_channels, affine=True, track_running_stats=True),
            nn.ReLU(inplace=True),
            nn.Conv2d(in_channels, in_channels, kernel_size=3, padding=1, bias=False),
            nn.InstanceNorm2d(in_channels, affine=True, track_running_stats=True)
        )

    def forward(self, x):
        return x + self.conv_block(x)
```

```python
class ConvBlock(nn.Module):

    def __init__(self, 
                 in_channels: int, 
                 out_channels: int, 
                 is_downsampling: bool, 
                 **kwargs):
        super(ConvBlock, self).__init__()

        if is_downsampling:
            self.conv_block = nn.Sequential(
                nn.Conv2d(in_channels, out_channels, padding_mode='reflect', **kwargs),
                nn.InstanceNorm2d(out_channels),
                nn.ReLU(inplace=True)
            )
        else:
            self.conv_block = nn.Sequential(
                nn.ConvTranspose2d(in_channels, out_channels, **kwargs),
                nn.InstanceNorm2d(out_channels),
                nn.ReLU(inplace=True)
            )

    def forward(self, x):
        return self.conv_block(x)
```

```python
class Generator(nn.Module):

    def __init__(self,
                 in_channels: int,
                 out_channels: int = 64,
                 num_blocks: int = 6) -> None:

        super(Generator, self).__init__()
        
        self.initial_layer = nn.Sequential(
            nn.Conv2d(in_channels, 
                      out_channels, 
                      kernel_size=7,
                      stride=1,
                      padding=3,
                      padding_mode='reflect'),
            nn.InstanceNorm2d(out_channels),
            nn.ReLU(inplace=True)
        )

        self.downsampling_layer = nn.ModuleList(
            [
                ConvBlock(out_channels, 
                          out_channels * 2,
                          is_downsampling = True,
                          kernel_size = 3,
                          stride=2,
                          padding=1),
                ConvBlock(out_channels * 2,
                          out_channels * 4,
                          is_downsampling = True,
                          kernel_size = 3,
                          stride = 2,
                          padding =1)
            ]
        )

        self.residual_layers = nn.Sequential(
            *[ResidualBlock(out_channels * 4) for _ in range(num_blocks)]
        )

        self.upsampling_layer = nn.ModuleList(
            [
                ConvBlock(out_channels * 4, 
                          out_channels * 2,
                          is_downsampling = False,
                          kernel_size = 3,
                          stride=2,
                          padding=1),
                ConvBlock(out_channels * 2,
                          out_channels,
                          is_downsampling = False,
                          kernel_size = 3,
                          stride = 2,
                          padding =1)
            ]
        )

        self.last_layer = nn.Conv2d(out_channels, 
                            in_channels, 
                            kernel_size=7,
                            stride=1,
                            padding=3,
                            padding_mode='reflect')
        
    def forward(self, x):
        x = self.initial_layer(x)
        for layer in self.downsampling_layer:
            x = layer(x)
        x = self.residual_layers(x)
        for layer in self.upsampling_layer:
            x = layer(x)
        x = self.last_layer(x)

        return torch.tanh(x)
```

__Now, we have the discriminator architecture__

![Alt text](../images/discriminator_architecture.png?raw=true "Discriminator Architecture")

```python
class Discriminator(nn.Module):

    def __init__(self, in_channels=3, out_features = 64):

        super(Discriminator, self).__init__()

        self.block1 = nn.Sequential(
            nn.Conv2d(in_channels, out_features, kernel_size=4, stride=2, padding=1, padding_mode='reflect'),
            nn.LeakyReLU(0.2, inplace=True)
        )

        self.block2 = nn.Sequential(
            nn.Conv2d(out_features, out_features * 2, kernel_size=4, stride=2, padding=1, padding_mode='reflect'),
            nn.LeakyReLU(0.2, inplace=True)
        )

        self.block3 = nn.Sequential(
            nn.Conv2d(out_features * 2, out_features * 4, kernel_size=4, stride=2, padding=1, padding_mode='reflect'),
            nn.LeakyReLU(0.2, inplace=True)
        )

        self.block4 = nn.Sequential(
            nn.Conv2d(out_features * 4, 1, kernel_size=4, stride=1, padding=1, padding_mode='reflect')
        )

    def forward(self, x):
        x = self.block1(x)
        x = self.block2(x)
        x = self.block3(x)
        x = self.block4(x)
        
        return torch.sigmoid(x)
```

## Creating the dataset loader

```python
class CycleGANDataset(Dataset):
    def __init__(self, root_dir_A, root_dir_B, transform=None):

        #set the directory path and the transform bool
        self.root_dir_A = root_dir_A
        self.root_dir_B = root_dir_B
        self.transform = transform

        #get the names of the files in the 2 folders
        self.image_files_A = [filename for filename in os.listdir(root_dir_A) if any(filename.endswith(extension) for extension in ['.jpg', '.png', '.jpeg'])]
        self.image_files_B = [filename for filename in os.listdir(root_dir_B) if any(filename.endswith(extension) for extension in ['.jpg', '.png', '.jpeg'])]

    def __len__(self):
        #return the min length of the two folders so that it iterates on the ones with minimum data
        return min(len(self.image_files_A), len(self.image_files_B))

    def __getitem__(self, idx):
        img_name_A = os.path.join(self.root_dir_A, self.image_files_A[idx % len(self.image_files_A)])
        img_name_B = os.path.join(self.root_dir_B, self.image_files_B[idx % len(self.image_files_B)])

        image_A = Image.open(img_name_A).convert('RGB')
        image_B = Image.open(img_name_B).convert('RGB')

        if self.transform:
            image_A = self.transform(image_A)
            image_B = self.transform(image_B)

        return {'A': image_A, 'B': image_B}

# Example usage:
data_transform = transforms.Compose([
    transforms.Resize((256, 256)),
    transforms.ToTensor(),
    transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))  # Normalize to [-1, 1]
])
```

```python
dataset = CycleGANDataset(root_dir_A='../data/Cyclegans/celebs', root_dir_B='../data/Cyclegans/comics', transform=data_transform)
dataloader = DataLoader(dataset, batch_size=6, shuffle=True)
```

```python
# Function to plot a batch of images
def plot_batch(batch, type):
    plt.figure(figsize=(10, 10))

    if type == "celebs":
        batch_type = "A"
    elif type == "comics":
        batch_type = "B"
    else:
        raise ValueError("Value of the type should be either celebs or comics!")

    for i in range(len(batch[batch_type])):
        plt.subplot(3, 2, i+1)
        plt.axis("off")
        plt.imshow((batch[batch_type][i].permute(1, 2, 0) + 1) / 2)  # Unnormalize and convert to numpy array
    plt.show()
```

```python
# Example usage:
batch = next(iter(dataloader))  # Get a batch of data from the DataLoader
plot_batch(batch, "celebs")
```

```python
#plotting the comics
plot_batch(batch, "comics")
```
